```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
X,Y=np.meshgrid(np.linspace(-1.57,1.57,100),np.linspace(-1.57,1.57,100))
Z=np.sin(2*X)+1.5*np.cos(Y)**4+2*np.cos(X+Y)
#plt.contour(X,Y,Z,50)
X=X[::5,::5]
Y=Y[::5,::5]
VX=2*np.cos(2*X)-2*np.sin(X+Y)
VY=-6*np.cos(Y)**3*np.sin(Y)-2*np.sin(X+Y)
plt.quiver(X,Y,VX,VY)
plt.show()
```

```python
I,S = np.meshgrid(np.linspace(0,1,20),np.linspace(0,1,20))
beta= 0.4
gamma = 0.2

Sp = (-1) * beta * S * I
Ip = beta*S*I - gamma*I
plt.quiver(S,I,Sp,Ip)

import random


for _ in range(4):
  x = random.random()
  print(x)
  sss = []
  iss = []
  s = x
  i = 1 - s

  print(i, s)
  while i >= 0.005:
    sss.append(s)
    iss.append(i)
    news =  (-1)*beta * s * i
    newi = (beta)*s*i - gamma*i
    s += news
    i += newi

  plt.plot(sss,iss)

plt.plot([0,1],[1,0])

plt.show()
```
